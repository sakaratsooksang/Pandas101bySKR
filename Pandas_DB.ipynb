```python
def connectDB(server):
    connection = pg.connect(f"host={server['host']} dbname={server['dbname']} user={server['user']} password={server['password']} port={server['port']}")
    return connection
def db2df(sql,connection):
    dataframe = psql.read_sql(sql, connection)
    return dataframe
```

# IS NULL / IS NOT NULL
NULL checking is done using the notnull() and isnull() methods.

```python
# SQL
sql = 'SELECT amphur_name, parcel_pincode FROM parcel_nonthaburi WHERE parcel_pincode IS NOT NULL LIMIT 100000'
df_sql = db2df(sql,connectDB(server))
df_sql
```

```python
# Pandas
df_pd = dataframe[['amphur_name','parcel_pincode']].loc[dataframe.parcel_pincode.isnull()]
df_pd
```

```python
# Pandas
df_pd = dataframe[['amphur_name','parcel_pincode']].loc[dataframe.parcel_pincode.notnull()]
df_pd
```

# **JOIN**
เนื่องจาก ไม่มีข้อมูลใน database ที่จะสามารถนำมา join กันได้ <br>
ดังนั้นจึงจะแสดงเพียงการนำ Dataframe มา Join กันใน pandas เท่านั้น

# INNER Join

```python
df1 = pd.DataFrame({"key": ["A", "B", "C", "D"], "value": np.random.randn(4)})
df2 = pd.DataFrame({"key": ["B", "D", "D", "E"], "value": np.random.randn(4)})
```

```python
df1
```

```python
df2
```

```python
# เลือก Column ที่ต้องการ Join = 'key'
pd.merge(df1, df2, on="key")
```

# LEFT OUTER JOIN

```python
pd.merge(df1, df2, on="key", how="left") 
#how เป็นการบอกว่าจะJoinอย่างไร
```

# RIGHT OUTER JOIN

```python
pd.merge(df1, df2, on="key", how="right")
```

# FULL OUTER JOIN

```python
pd.merge(df1, df2, on="key", how="outer")
```

# **Geodatabase To GeoDataFrame**

```python
import geopandas as gpd
```

```python
def connectDB(server):
    connection = pg.connect(f"host={server['host']} dbname={server['dbname']} user={server['user']} password={server['password']} port={server['port']}")
    return connection
def gdb2gdf(sql,connection,geometry_column,epsg):
    geodataframe = gpd.GeoDataFrame.from_postgis(sql, connection, geom_col=geometry_column,crs=f'EPSG:{epsg}')
    return geodataframe
```

```python
sql='SELECT amphur_name,the_geom  FROM parcel_nonthaburi LIMIT 10000;'
gdf = gdb2gdf(sql,connectDB(server),'the_geom','24047')
```
