# การเลือกแถวแบบมีเงื่อนไข
# ถ้ามากกว่า1เงื่อนไขต้องใส่ .loc

```python
# วันใดบ้างที่มีอุณหภูมิ <=20
[df.Temperature<=20]
```

```python
df[df.Temperature<=20]
```

```python
# วันใดบ้างที่มีอุณหภูมิ >=35 และฝนตก
df.loc[(df.Event=='ฝนตก') & (df.Temperature>=35)]
#### ถ้ามีมากกว่า 1 เงื่อนไขต้องใส่ loc
```

```python
#วันใดบ้างที่มีอุณหภูมิ <=18 หรือเมฆมาก
df.loc[(df.Event=='เมฆมาก') | (df.Temperature<=18)]
```

# ISIN(List) FUNCTION
เช็คว่ามีค่านี้อยู่ในตรงไหนบ้างหรือเปล่า
ใช้แทน OR

```python
df.loc[(df.Temperature.isin([18,20,25,35]))]
# หาดูว่ามีวันใดบ้างที่ temp = 18 or 20 or 25 or 35 deg
```

```python
df.loc[(df.Temperature.isin([18,20,25,35])) & df.Event.isin(['แดดร้อน']) ]
```

```python
# หรือสามารถนำมาเลือกแบบเขียนต่อท้ายได้เช่นกัน
df.loc[df.Temperature.isin([25,30])][df.Event=='แดดร้อน']
```
